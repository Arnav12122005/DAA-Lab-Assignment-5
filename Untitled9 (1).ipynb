```python
"""
Refactored delivery optimization notebook as a single Python module.
Features:
 - Problem data (locations, distances, parcels, vehicle capacity)
 - TSP solutions: brute-force and Held-Karp (DP)
 - Shortest paths (Dijkstra)
 - MST (Prim)
 - Greedy knapsack for parcel selection
 - Route time feasibility w/ delivery time windows
 - Simple visualization (matplotlib)

Usage: run as a script or import functions for testing.
"""

import itertools
import math
from functools import lru_cache
import heapq
import random
import matplotlib.pyplot as plt

locations = ["Warehouse", "C1", "C2", "C3"]

distance_matrix = [
    [0, 4, 8, 6],
    [4, 0, 5, 7],
    [8, 5, 0, 3],
    [6, 7, 3, 0],
]
parcels = {
    "C1": {"value": 50, "time": (9, 12), "weight": 10},
    "C2": {"value": 60, "time": (10, 13), "weight": 20},
    "C3": {"value": 40, "time": (11, 14), "weight": 15},
}
vehicle_capacity = 30


def index_of(loc_name):
    return locations.index(loc_name)


def tsp_bruteforce(dist_mat, locs=None):
    """Return best route and cost by checking all permutations (start/end at 0)."""
    if locs is None:
        locs = locations
    n = len(dist_mat)
    indices = list(range(1, n))
    best_cost = math.inf
    best_route = None
    for perm in itertools.permutations(indices):
        cost = dist_mat[0][perm[0]]
        for i in range(len(perm) - 1):
            cost += dist_mat[perm[i]][perm[i + 1]]
        cost += dist_mat[perm[-1]][0]
        if cost < best_cost:
            best_cost = cost
            best_route = (0,) + perm + (0,)
    return [locs[i] for i in best_route], best_cost


def held_karp(dist_mat):
    """Held-Karp dynamic programming solution for TSP. Returns optimal cost and parent table."""
    n = len(dist_mat)

    C = {}
    for k in range(1, n):
        C[(1 << (k - 1), k)] = (dist_mat[0][k], 0)
    for subset_size in range(2, n):
        for subset in itertools.combinations(range(1, n), subset_size):
            bits = 0
            for bit in subset:
                bits |= 1 << (bit - 1)
            for k in subset:
                prev_bits = bits & ~(1 << (k - 1))
                candidates = []
                for m in subset:
                    if m == k:
                        continue
                    if (prev_bits, m) in C:
                        candidates.append((C[(prev_bits, m)][0] + dist_mat[m][k], m))
                if candidates:
                    C[(bits, k)] = min(candidates)
    bits_all = (1 << (n - 1)) - 1
    res = []
    for k in range(1, n):
        if (bits_all, k) in C:
            res.append((C[(bits_all, k)][0] + dist_mat[k][0], k))
    if not res:
        return None, None
    opt_cost, parent_k = min(res)
    return opt_cost, C


def dijkstra(dist_mat, src=0):
    n = len(dist_mat)
    dist = [math.inf] * n
    parent = [-1] * n
    dist[src] = 0
    pq = [(0, src)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v in range(n):
            w = dist_mat[u][v]
            if w > 0 and dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                parent[v] = u
                heapq.heappush(pq, (dist[v], v))
    return dist, parent


def prim_mst(dist_mat):
    n = len(dist_mat)
    in_mst = [False] * n
    key = [math.inf] * n
    parent = [-1] * n
    key[0] = 0
    for _ in range(n):
        u = min((key[i], i) for i in range(n) if not in_mst[i])[1]
        in_mst[u] = True
        for v in range(n):
            w = dist_mat[u][v]
            if w > 0 and not in_mst[v] and w < key[v]:
                key[v] = w
                parent[v] = u
    edges = []
    total = 0
    for i in range(1, n):
        edges.append((parent[i], i, dist_mat[parent[i]][i]))
        total += dist_mat[parent[i]][i]
    return edges, total


def greedy_select(parcels_dict, capacity):
    items = []
    for loc, info in parcels_dict.items():
        ratio = info['value'] / info['weight']
        items.append((loc, info['value'], info['weight'], ratio))
    items.sort(key=lambda x: x[3], reverse=True)
    chosen = []
    total_weight = 0
    total_value = 0
    for loc, val, w, _ in items:
        if total_weight + w <= capacity:
            chosen.append(loc)
            total_weight += w
            total_value += val
    return chosen, total_value, total_weight



def feasible_route_time(order_indices, start_time=9, dist_mat=None, parcels_map=None):
    """Check if visiting the given indices in order meets parcel time windows.
    order_indices are indices (integers) into `locations` (not including 0 for warehouse unless included).
    start_time is numeric (e.g., 9 for 9am); we add travel time equal to distances.
    """
    if dist_mat is None:
        dist_mat = distance_matrix
    if parcels_map is None:
        parcels_map = parcels
    t = start_time
    pos = 0
    for idx in order_indices:
        t += dist_mat[pos][idx]
        earliest, latest = parcels_map[locations[idx]]['time']

        if t > latest:
            return False

        if t < earliest:
            t = earliest
        pos = idx
    return True


def reconstruct_held_karp_path(C, dist_mat):
    n = len(dist_mat)
    bits_all = (1 << (n - 1)) - 1

    res = []
    for k in range(1, n):
        if (bits_all, k) in C:
            res.append((C[(bits_all, k)][0] + dist_mat[k][0], k))
    opt, parent_k = min(res)
    path = [0]
    bits = bits_all
    k = parent_k
    stack = []
    while bits:
        stack.append(k)
        _, parent = C[(bits, k)]
        bits &= ~(1 << (k - 1))
        k = parent
    path.extend(reversed(stack))
    path.append(0)
    return [locations[i] for i in path]


def plot_route(route_indices, title="Route"):

    n = len(route_indices)
    angles = [2 * math.pi * i / n for i in range(n)]
    xs = [math.cos(a) for a in angles]
    ys = [math.sin(a) for a in angles]
    plt.figure(figsize=(6, 6))
    for i, lab in enumerate(route_indices):
        plt.scatter(xs[i], ys[i])
        plt.text(xs[i] * 1.05, ys[i] * 1.05, lab)
    for i in range(len(route_indices) - 1):
        plt.plot([xs[i], xs[i + 1]], [ys[i], ys[i + 1]])
    plt.title(title)
    plt.axis('off')
    plt.show()


if __name__ == "__main__":
    print("Locations:", locations)
    print("Distance matrix:")
    for row in distance_matrix:
        print(row)
    print("Parcels:", parcels)
    print("Capacity:", vehicle_capacity)


    rec_cost, C = held_karp(distance_matrix)
    print("Held-Karp optimal tour cost:", rec_cost)

    route_bf, cost_bf = tsp_bruteforce(distance_matrix)
    print("Brute-force TSP route:", route_bf, "cost:", cost_bf)

    chosen, val_sum, wt_sum = greedy_select(parcels, vehicle_capacity)
    print("Greedy chosen parcels:", chosen, "Total value:", val_sum, "Total weight:", wt_sum)


    order = [index_of('C1'), index_of('C2'), index_of('C3')]
    print("Order", [locations[i] for i in order], "feasible?", feasible_route_time(order, start_time=8))

    dist_from_wh, parent = dijkstra(distance_matrix, 0)
    print("Shortest distances from warehouse:", dist_from_wh)

    mst_edges, mst_cost = prim_mst(distance_matrix)
    print("MST edges:", mst_edges, "cost:", mst_cost)


    try:
        plot_route(route_bf, title=f"TSP route (cost {cost_bf})")
    except Exception:
        pass
```
